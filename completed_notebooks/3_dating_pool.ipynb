```python
plus_divorced = {
    "B12002_002E": "total_males",
    "B12002_009E": "35_39",
    "B12002_010E": "40_44",
    "B12002_011E": "45_49",
}
```

```python
# searching for cities in the census
# use ACS1 to find only places with populations > 65,000
ced.download(
    "acs/acs1",
    2023,
    download_variables=["NAME"],
    state=NC,
    place="*",
)
```

```python
pool = ced.download(
    "acs/acs5",
    2023,
    download_variables = plus_divorced.keys(),
    state = NC,
    place = "02140"
).rename(
    columns = plus_divorced
)

pool.head()
```
